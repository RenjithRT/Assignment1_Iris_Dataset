# Iris Data Set

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings("ignore") # to avoid warning messages
```

### 1. Read the dataset

```python
iris_data = pd.read_excel("iris.xls")
```

```python
iris_data
```

- The iris dataset contains the following data
    - 50 samples of 3 different species of iris (150 samples total)
    - Measurements: sepal length, sepal width, petal length, petal width

### 2. Display the columns in the dataset

```python
# The summary of the iris dataset
iris_data.info()
```

```python
iris_data.columns
```

- The columns of the data: SL(sepal length), SW(sepal width), PL(petal length), PW(petal width) and Classification

### 3. Calculate the mean of each column of the dataset

```python
# The Statistical summary of iris dataset
iris_data.describe()
```

Here we can see the complete statistical summary of count, mean, statndard deviation, minimum value of each column also maximum, 25%, 50% and 75% percentile.

```python
iris_data.mean()
```

The mean of SL(sepal length), SW(sepal width), PL(petal length) and PW(petal width) are 5.84, 3.05, 3.75 and 1.19 respectively.

### 4. Check for the null values present in the dataset

```python
# Checking any null values present in the dataset
iris_data.isna() # otherwise we can use iris_data.isnull()
```

```python
# Calculating the null values present in each columns in the dataset
iris_data.isna().sum()
```

### 5. Perform meaningful visualizations using the dataset. Bring at least 3 visualizations

In our dataset, there is no missing values present, so we can go for visualizations.

```python
iris_data['Classification'].value_counts()
```

- There are 50 observations of each classification (setosa, versicolor, virginica).

#### Using Pairplot

```python
sns.pairplot(iris_data, hue = 'Classification')
```

- In the above graph, the features in a pair plot, it is clear that the relationship between pairs of features of a iris-setosa (in blue) is distinctly different from those of the other two classifications.
- There is some overlap in the pairwise relationships of the other two classifications, iris-versicolor (orange) and iris-virginica (green).

#### Using Violinplot

```python
sns.violinplot(x = 'SL', y = 'Classification', data = iris_data, inner = 'quartile')
```

```python
sns.violinplot(x = 'SW', y = 'Classification', data = iris_data, inner = 'quartile')
```

```python
sns.violinplot(x = 'PL', y = 'Classification', data = iris_data, inner = 'quartile')
```

```python
sns.violinplot(x = 'PW', y = 'Classification', data = iris_data, inner = 'quartile')
```

In the above four violin plot, you can see the argument inner we passed quartile, so that's the reason, inside violin plot its just showing the similar plot like boxplot. Also we can say that each continous column is classified with categorical column. 

#### Using Heatmap

```python
sns.heatmap(iris_data.corr(), annot = True)
```

In this heatmap, we can identifies that correlation between featues, with high corelation between Petal Length(PL), Petal Width(PW) and Sepal Length(SL).

#### USing Histogram

```python
iris_data[['SL','SW','PL','PW']].plot.hist()
```

The above histogram shows the frequency distribution of Sepal Length(SL), Sepal Width(SW), Petal Length(PL), Petal Width(PW)

#### Using KDEplot (Kernel Base Density)

```python
iris_data[['SL','SW','PL','PW']].plot.kde()
```

In the above kdeplot, it depicts the probability distribution at different values in a countinous variables, includes SL(sepal length), SW(sepal width), PL(petal length), PW(petal width)

#### Using Strip Plot

```python
sns.stripplot(x = 'Classification', y = 'SL', data = iris_data, jitter=True)
```

```python
sns.stripplot(x = 'Classification', y = 'SW', data = iris_data, jitter=True)
```

```python
sns.stripplot(x = 'Classification', y = 'PL', data = iris_data, jitter=True)
```

```python
sns.stripplot(x = 'Classification', y = 'PW', data = iris_data, jitter=True)
```

In the above four strip plots, we applied a small random jitter value to each point such that the seperation between points becomes clearer. 
- The x axis represents the categorical variable (Classification) and y axis represents the continuous varible (Sepal Length(SL), Sepal Width(SW), Petal Length(PL), Petal Width(PW))

#### Using Swarm Plot

```python
sns.swarmplot(x = 'Classification', y = 'SL', data = iris_data)
```

```python
sns.swarmplot(x = 'Classification', y = 'SW', data = iris_data)
```

```python
sns.swarmplot(x = 'Classification', y = 'PL', data = iris_data)
```

```python
sns.swarmplot(x = 'Classification', y = 'PW', data = iris_data)
```

In the above four swarm plots, similar to strip plot, yet the locations of points are adjusted automatically to avoid overlap even if the jitter value is not applied. So it clear that we comparing swarm plot with strip plot, where we can see the all points are located in clearly.
